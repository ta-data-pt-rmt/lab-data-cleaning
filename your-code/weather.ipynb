## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
weather = pd.read_csv('weather-raw.csv')
weather.columns = weather.columns.str.replace("d", "")
display(weather.sample(10))
```

```python
weather = weather.fillna(0)
display(weather.sample(10))
display(weather.dtypes)
```

```python
weather.columns = weather.columns.str.replace("d", "")

weather_melt = weather.melt(id_vars = ['i','year','month','element'],var_name='day')

weather_melt = weather_melt.pivot_table('value', ['i', 'year', 'month', 'day'], 'element').reset_index()
```

```python
display(weather_melt.head(10))
```

```python
weather_melt['Datetime'] = weather_melt[['year','month','day']].astype(str).apply('-'.join,1)
#weather_melt['Datetime'] = pd.to_datetime(weather_melt['Datetime'])
```
