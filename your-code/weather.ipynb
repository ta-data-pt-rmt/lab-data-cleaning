## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
MX17004 = pd.read_csv('weather-raw.csv')
```

```python
MX17004
```

```python
MX17004 = pd.melt(MX17004, id_vars = ['id', 'year', 'month', 'element'], var_name = 'days', value_name = 'degrees')
MX17004.sample(5)
```

```python
MX17004 = MX17004.pivot_table(index = ['id','year', 'month','days'], columns = 'element', values = 'degrees').reset_index()

MX17004.head()
```
