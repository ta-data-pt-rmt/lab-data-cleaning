## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather_df = pd.read_csv('weather-raw.csv')
weather_df
```

```python
columns_names = weather_df.columns

for i in columns_names:
    print(i)
```

```python
weather_df2=weather_df.melt(id_vars=["id","year","month","element"],var_name=["day"],value_name="temp")
weather_df2
weather_df2.update(pd.DataFrame({"day":[day[1:] for day in weather_df2["day"]]}))
weather_df2.sample(5)
```

```python
weather_df3=(weather_df2.pivot_table(index = ["year","month","day","id"], columns = "element", values = "temp")
       .reset_index().rename_axis(None, axis = 1))
weather_df3
```

```python
weather_df4=(weather_df3.assign(date = lambda x: x.day.astype("str") +"/"+ x.month.astype("str").str.zfill(2) +"/"+ x.year.astype("str").str.zfill(2))
            .drop(["year", "month","day"],axis=1))
# weather_df4['date'] =  pd.to_datetime(weather_df4['date'], dayfirst=True, format='%d/%m/%Y')
weather_df4
```

```python
weather_df5=(weather_df4.filter(["date", "tmax","tmin"]).groupby("date").aggregate("mean").reset_index())
weather_df5
```
