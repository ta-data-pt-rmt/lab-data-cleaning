#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users = pd.read_csv("./datasets/users_table.csv")
users.head()
```

```python
display(users.dtypes)
```

#### 3. Rename Id column to userId

```python
users = users.rename(columns = {'Id':'userId'})
print(users.columns)
```

```python
users['userId'] = users['userId'].astype(float)
display(users.dtypes)
```

#### 4. Import posts table:

```python
posts_table = pd.read_csv("./datasets/posts_table.csv")
posts_table.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_table = posts_table.rename(columns = {'Id':'postId','OwnerUserId':'userId' })
print(posts_table.columns)
```

```python
posts_table['ViewCount'] = posts_table['ViewCount'].fillna(0)
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users2 = users[['userId', 'Reputation','Views','UpVotes','DownVotes']]
posts_table2 = posts_table[['postId', 'Score','userId','ViewCount','CommentCount']]

display(users2.shape)
display(users2.dtypes)
users2.head()
```

```python
display(posts_table2.shape)
display(posts_table2.dtypes)
posts_table2.head()
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_data = pd.merge(left = users2,
                       right = posts_table2,
                       how = 'inner', 
                       left_on = "userId", 
                       right_on= "userId")
display(merged_data.shape)
display(merged_data.dtypes)
merged_data.head()
```

```python
merged_data.describe()
```

```python
merged_data.isnull().sum()
```
