#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users = pd.read_csv("users_table.csv")
users.head()

```

#### 3. Rename Id column to userId

```python
users.rename(columns={"Id": "userId"}, inplace=True)
users.head()
```

#### 4. Import posts table:

```python
posts = pd.read_csv("posts_table.csv")
posts.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename(columns={"Id": "postId"}, inplace=True)
posts.rename(columns={"OwnerUserId": "userId"}, inplace=True)
posts.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_columns = users[['userId','Reputation','Views','UpVotes','DownVotes']]
posts_columns = posts[['postId','Score','userId','ViewCount','CommentCount']]
```

#### 7. Merge both dataframes, users and posts.
You will need to make a [merge](https://urldefense.com/v3/__https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html__;!!AcCyiFYNC0XOnw!lfCFXE5uGCv9qjrLdZY-DCos1fSZs94iRmGBBKQPS046lVdCRPm33OgzkPghT7p2wPqquqJwoGerp4s2nRkGJQ$ ) of posts and users dataframes.

```python
merged_df = posts.merge(users, left_on='postId', right_on='userId')
merged_df.head()
```
