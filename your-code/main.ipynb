#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users_table = pd.read_csv("datasets/users_table.csv")
display(users_table)
```

#### 3. Rename Id column to userId

```python
users_table.rename(columns={"Id": "userId"}, inplace=True)

display(users_table)
```

#### 4. Import posts table:

```python
posts = pd.read_csv("datasets/posts_table.csv")
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
# Rename the "Id" column to "postId" and "OwnerUserId" to "userId" in the posts DataFrame
posts.rename(columns={"Id": "postId", "OwnerUserId": "userId"}, inplace=True)
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
# Define a new DataFrame for users with selected columns
selected_user_columns = ["userId", "Reputation", "Views", "UpVotes", "DownVotes"]
users_selected = users_table[selected_user_columns]

# Define a new DataFrame for posts with selected columns
selected_post_columns = ["postId", "Score", "userId", "ViewCount", "CommentCount"]
posts_selected = posts[selected_post_columns]

display(posts_selected)
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_data = users_selected.merge(posts_selected, on="userId")

display(merged_data)
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
# Find the number of missing values in the merged DataFrame
missing_values = merged_data.isnull().sum()

# Filter columns with missing values
columns_with_missing_values = missing_values[missing_values > 0]

# Display the number of missing values and columns with missing values
print("Number of missing values in the merged DataFrame:")
print(missing_values)

print("\nColumns with missing values:")
print(columns_with_missing_values)
```
